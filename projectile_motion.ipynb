```python
# projectile with air resistance related to v^2

import numpy as np
import matplotlib.pyplot as plt

g = 9.8
v0 = 4
x0 = 0 
y0 = 0
D = 0.95

v0x = v0*(np.cos(np.pi/4))
v0y = v0*(np.sin(np.pi/4))
v = (v0x**2 + v0y**2)**0.5
ax = -D*v*v0x
ay = -g - D*v*v0y

height = []
xRange = []

deltaT=0.01

yCutOff = y0

while y0>=yCutOff:
    height.append(y0)
    xRange.append(x0)
    
    y0+=v0y*deltaT + 0.5*ay*deltaT**2
    v0y+=ay*deltaT
    x0+= v0x*deltaT + 0.5*ax*deltaT**2
    v0x+=ax*deltaT
    
plt.plot (xRange,height , 'r')
```
